```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagen = cv2.imread('imagenes/minion.jpg')
```

```python
plt.imshow(imagen)
```

```python
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
plt.imshow(imagen)
```

```python
imagen.shape
```

```python
imagen2 = cv2.resize(imagen, (600,500))
```

```python
plt.imshow(imagen2)
```

```python
ratio_ancho = 0.5
ratio_alto = 0.5
```

```python
imagen3 = cv2.resize(imagen2, (0,0), imagen2, ratio_ancho, ratio_alto)
```

```python
plt.imshow(imagen3)
```

```python
imagen3.shape
```

```python
imagen4= cv2.flip(imagen3, 1)
```

```python
plt.imshow(imagen4)
```

```python
imagen4= cv2.flip(imagen3, 0)
plt.imshow(imagen4)
```

```python
imagen5 = cv2.cvtColor(imagen4, cv2.COLOR_RGB2BGR)
```

```python
cv2.imwrite('/imagenes/minion_al_reves.jpg', imagen5)
```
