```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagen = cv2.imread('imagenes/sudoku.jpg', 0)
plt.imshow(imagen, cmap='gray')
```

**Identificar contornos**

```python
sobelx = cv2.Sobel(imagen, cv2.CV_64F, 1, 0, ksize=5) # 1 detecta verticales, 0 horizontales
plt.imshow(sobelx, cmap='gray')
```

```python
sobely = cv2.Sobel(imagen, cv2.CV_64F, 0,1, ksize=5) # 1 detecta verticales, 0 horizontales
plt.imshow(sobely, cmap='gray')
```

```python
imagen2 = cv2.addWeighted(src1=sobelx, alpha = 0.5, src2=sobely, beta=0.5, gamma=0)
plt.imshow(imagen2, cmap='gray')
```

```python
# Prueba mía
sobelz = cv2.Sobel(imagen, cv2.CV_64F, 1,1, ksize=5) # 1 detecta verticales, 0 horizontales
plt.imshow(sobelz, cmap='gray')
```

**Método Laplacian**

```python
laplacian = cv2.Laplacian(imagen, cv2.CV_64F)
plt.imshow(laplacian, cmap='gray')
```
