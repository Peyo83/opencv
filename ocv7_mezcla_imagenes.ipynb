```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagen1 = cv2.imread('imagenes/minion.jpg')
imagen1 = cv2.cvtColor(imagen1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagen1)
```

```python
imagen2 = cv2.imread('imagenes/logo-redaitcursos.jpg')
imagen2 = cv2.cvtColor(imagen2, cv2.COLOR_BGR2RGB)
plt.imshow(imagen2)
```

```python
imagen1.shape
```

```python
imagen2.shape
```

```python
imagen2 = cv2.resize(imagen2, (imagen1.shape[1], 250))
```

```python
plt.imshow(imagen2)
```

```python
parte_imagen1 = imagen1[:250]
```

```python
plt.imshow(parte_imagen1)
```

```python
parte_imagen_mezcla = cv2.addWeighted(src1=parte_imagen1, alpha = 0.5, src2 = imagen2, beta=0.5, gamma=0)
```

```python
plt.imshow(parte_imagen_mezcla)
```

```python
imagen1[:250] = parte_imagen_mezcla
plt.imshow(imagen1)
```
