```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagen = cv2.imread('imagenes/perro.jpg')
```

```python
plt.imshow(imagen)
```

```python
imagen = cv2.imread('imagenes/minion.jpg', 0)
```

```python
plt.imshow(imagen, cmap='gray')
```

```python
mitad, imagen = cv2.threshold(imagen, 255/2, 255, cv2.THRESH_BINARY)
plt.imshow(imagen, cmap='gray')
```

```python
mitad, imagen = cv2.threshold(imagen, 255/2, 255, cv2.THRESH_BINARY_INV)
plt.imshow(imagen, cmap='gray')
```

```python
figura = plt.figure(figsize=(20,20))
lienzo = figura.add_subplot(111)
lienzo.imshow(imagen)
```

```python
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
plt.imshow(imagen)
```

```python
figura = plt.figure(figsize=(20,15))
lienzo = figura.add_subplot(111)
lienzo.imshow(imagen)
```
