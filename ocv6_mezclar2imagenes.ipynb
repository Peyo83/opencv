```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagen = cv2.imread('imagenes/minion.jpg')
imagen1 = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagen1)
```

```python
imagen = cv2.imread('imagenes/marca-agua-redaitcurso.jpg')
imagen2 = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagen2)
```

```python
imagen1.shape, imagen2.shape
```

```python
imagen1 = cv2.resize(imagen1, (1200, 1200))
imagen2 = cv2.resize(imagen2, (1200, 1200))
```

```python
plt.imshow(imagen1)
```

```python
plt.imshow(imagen2)
```

```python
imagen_mezcla = cv2.addWeighted(src1=imagen1, alpha = 0.5, src2 = imagen2, beta = 0.5, gamma = 0)
```

```python
plt.imshow(imagen_mezcla)
```

```python
imagen_mezcla = cv2.addWeighted(src1=imagen1, alpha = 1, src2 = imagen2, beta = 0.7, gamma = 0)
```

```python
plt.imshow(imagen_mezcla)
```
