```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagen = cv2.imread('imagenes/minion.jpg')
plt.imshow(imagen)              
```

cambio escala de colores

```python
imagen2 = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
plt.imshow(imagen2)
```

```python
imagen2 = cv2.cvtColor(imagen, cv2.COLOR_BGR2HSV)
plt.imshow(imagen2)
```

```python
imagen2 = cv2.cvtColor(imagen, cv2.COLOR_BGR2HLS)
plt.imshow(imagen2)
```
